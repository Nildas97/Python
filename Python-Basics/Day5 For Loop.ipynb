### from this tuple extract the value 3?

```python
t = ("nil", "das", [1,2,3,4,5], (5,4,3,2,1), {3:3, "key1": "nil", "key2": "das"})
```

```python
for i in t:
    print(i)
```

```python
for i in t:
    if type(i) == list or type(i) == tuple:
        for j in i:
            if j == 3:
                print(j)
    if type(i) == dict:
        for k, v in i.items():
            if k == 3 or v == 3:
                print(k)
                print(v)
```
