```python
# q10 : write a code to trigger alarm for you at scheduled time 
import time
import winsound

def set_alarm(alarm_time):
    while True:
        current_time = time.strftime("%H:%M:%S", time.localtime())
        if current_time == alarm_time:
            print("Wake up!")
            duration = 1000  # milliseconds
            freq = 550  # Hz
            winsound.Beep(freq, duration)
            break

        time.sleep(1)

# Set the desired alarm time (24-hour format)
alarm_time = "23:35:00"  # Example: 8:00 AM

set_alarm(alarm_time)
```

```python
# q10 : write a code to trigger alarm for you at scheduled time 
import datetime
import time
import winsound

def set_alarm(alarm_time):
    current_time = datetime.datetime.now().strftime("%H:%M:%S")
    print("Current Time:", current_time)

    while current_time < alarm_time:
        time.sleep(1)
        current_time = datetime.datetime.now().strftime("%H:%M:%S")

    print("Alarm Time:", alarm_time)
    print("Wake up!")
    
    # Play alarm sound
    duration = 1000  # milliseconds
    freq = 550  # Hz
    winsound.Beep(freq, duration)

# Set the desired alarm time (24-hour format)
alarm_time = "23:35:00"  # Example: 8:00 AM


set_alarm(alarm_time)
```

```python
# q16 : Write a fun which will take input as a dict and give me out as a list of all the values even in case of 2 level nesting it should work.
def dic_to_list(d):
    l = []
    for i in d.values():
        if type(i) != dict:
            l.append(i)
        if type(i) == dict:
            for j in i.values():
                l.append(j)
    return l
```

```python
from PIL import Image

def show_image(file_path):
    try:
        image = Image.open(file_path)
        image.show()
    except FileNotFoundError:
        print("Image file not found.")
    except Exception as e:
        print("An error occurred:", str(e))

# Provide the file path of the image
file_path = "path_to_your_image.jpg"  # Replace with the actual file path

show_image(file_path)
```

```python
import cv2

def show_image(file_path):
    try:
        image = cv2.imread(file_path)
        cv2.imshow("Image", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    except FileNotFoundError:
        print("Image file not found.")
    except Exception as e:
        print("An error occurred:", str(e))

# Provide the file path of the image
file_path = "E:/Samsung Flow/Light Trails.png"  # Replace with the actual file path

show_image(file_path)
```
