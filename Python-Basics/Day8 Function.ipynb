### A prime number is a natural number greater than 1 that is divisible only by 1 and itself, without any other divisors. In other words, a prime number has exactly two distinct positive divisors: 1 and the number itself.

### To determine if a number is prime or not in a Python program, you can use the following logic:

### * Start with the number you want to check.<br> * If the number is less than 2, it is not prime.<br> * Iterate from 2 up to the square root of the number (inclusive).<br> * Check if the number is divisible evenly by any of the iterated values. If so, it is not prime.<br> * If the number is not divisible evenly by any of the iterated values, it is prime.

```python
#method 1
lower = 1
upper = 10

print("Prime numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
   # all prime numbers are greater than 1
   if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                print(num)
                break
        else:
            print(num)
```

```python
#method 2
def is_prime(num):
    if num < 2:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True

# Find and print prime numbers between 1 and 1000
for number in range(1, 1001):
    if is_prime(number):
        print(number)
```

```python
# q1 : Try to print a prime number in between 1 to 1000
def prime_no(num):
    if num < 2:
        return False
    for i in range(2,num):
        if num % i == 0:
            return False
    return True
```

```python
for number in range(1,1001):
    if prime_no(number):
        print(number)
```

```python
# q2 : Try to write a function which is equivalent to print function in python
def my_print(*args, sep=' ', end='\n'):
    output = sep.join(str(arg) for arg in args)
    print(output, end=end)
```
