### OR , AND & NOT

### OR

```python
a = 10
if a < 10 or a == 10:
    print("return the output")
```

```python
a = "nil"
if a == "nil" or a == "Nil":
    print("return the output")
```

### reason for both code - in the 'or' operator if any of the parameter are true then output will return.<br> if both operator are true output will return else not.

# Loop preview

### for loop

```python
l = [1,2,3,4,5,6,"nildas", 234.45]
```

```python
l
```

### forward loop result

```python
for i in l:
    print(i)
```

### backward loop result

```python
for i in l[::-1]:
    print(i)
```

```python
for i in l:
    print(i+1)
```

### here the loop starts from 1 and then according to the loop it will add 1. by following this it will go till 6+1 i.e. 7 after that string concatenation is not possible so loop gets break.

```python
for i in l:
    if type(i) == int or type(i) == float:
        print(i + 1)
    elif type(i) == str:
        print(i + str(1))
```

```python
l
```

### appending items into a empty list

```python
l1 = []
for i in l:
    if type(i) == int:
        l1.append(i)
```
